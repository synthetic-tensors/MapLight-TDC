```python
#df.status.value_counts()['Completed']
#statuses = ['Completed','Withdrawn','Terminated','Suspended']
def get_statuses(x):
    counts = x.value_counts()
    c = [counts[status] if status in counts.keys() else 0 for status in statuses ]
    return c
```

```python
df_status = df[df.status.isin(
    ['Completed','Withdrawn','Terminated','Suspended']
)].groupby('canonical_smiles').status.agg(lambda x: get_statuses(x))
```

```python
df_res = pd.DataFrame(df_status.to_list(),index = df_status.index, columns=statuses)
df_res['N'] = df_res['Withdrawn']+df_res['Terminated']+df_res['Suspended']
df_res['P'] = df_res['Completed']
```
